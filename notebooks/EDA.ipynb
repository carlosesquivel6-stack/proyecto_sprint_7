# Análisis Exploratorio de Datos (EDA)
## Dataset: Anuncios de venta de vehículos en EE.UU. (`vehicles_us.csv`)

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

## 1. Carga de datos

```python
df = pd.read_csv('../vehicles_us.csv')
print('Forma del dataset:', df.shape)
df.head()
```

## 2. Información general

```python
df.info()
```

```python
df.describe()
```

## 3. Valores ausentes

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
pd.DataFrame({'Valores ausentes': missing, '% del total': missing_pct})
```

## 4. Distribución de precios

```python
fig = px.histogram(df, x='price', nbins=50, title='Distribución de precios de vehículos')
fig.update_layout(xaxis_title='Precio (USD)', yaxis_title='Cantidad de anuncios')
fig.show()
```

## 5. Distribución del odómetro

```python
fig = px.histogram(df, x='odometer', nbins=50, title='Distribución del odómetro')
fig.update_layout(xaxis_title='Odómetro (millas)', yaxis_title='Cantidad de anuncios')
fig.show()
```

## 6. Relación entre odómetro y precio

```python
df_clean = df[(df['price'] > 500) & (df['price'] < 100000) & (df['odometer'] < 500000)]
fig = px.scatter(
    df_clean, x='odometer', y='price', color='condition',
    title='Relación entre Odómetro y Precio por Condición',
    opacity=0.4
)
fig.update_layout(xaxis_title='Odómetro (millas)', yaxis_title='Precio (USD)')
fig.show()
```

## 7. Distribución por tipo de vehículo

```python
type_counts = df['type'].value_counts().reset_index()
type_counts.columns = ['type', 'count']
fig = px.bar(type_counts, x='type', y='count', title='Cantidad de anuncios por tipo de vehículo')
fig.update_layout(xaxis_title='Tipo', yaxis_title='Cantidad')
fig.show()
```

## 8. Distribución por condición del vehículo

```python
condition_counts = df['condition'].value_counts().reset_index()
condition_counts.columns = ['condition', 'count']
fig = px.pie(condition_counts, names='condition', values='count', title='Distribución por condición')
fig.show()
```

## 9. Precio mediano por condición

```python
avg_price = df.groupby('condition')['price'].median().reset_index()
avg_price.columns = ['condition', 'precio_mediano']
avg_price = avg_price.sort_values('precio_mediano', ascending=False)
fig = px.bar(avg_price, x='condition', y='precio_mediano', title='Precio mediano por condición del vehículo')
fig.update_layout(xaxis_title='Condición', yaxis_title='Precio mediano (USD)')
fig.show()
```

## 10. Conclusiones

- El dataset contiene **51 525 anuncios** de venta de vehículos con 13 columnas.
- Las columnas con más valores ausentes son `paint_color`, `cylinders`, `is_4wd` y `model_year`.
- La mayoría de los precios se concentran entre \$1 000 y \$30 000 USD.
- Existe una correlación negativa entre el odómetro y el precio: a mayor kilometraje, menor precio.
- Los tipos de vehículo más anunciados son `SUV`, `sedan` y `pickup`.
- Los vehículos en condición `new` y `like new` tienen los precios medianos más altos.
